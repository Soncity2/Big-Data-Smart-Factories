```python
from kafka import KafkaProducer
import pandas as pd
import time
import json
import threading

# Kafka configuration
KAFKA_TOPIC = 'Company_B'
KAFKA_BROKER = 'localhost:9092'

# Load dataset
df = pd.read_csv('company_B.csv')

# Get unique assets
unique_assets = df['asset'].unique()

# Initialize Kafka Producers
producers = {asset: KafkaProducer(
    bootstrap_servers=KAFKA_BROKER,
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
) for asset in unique_assets}

# Function to send data for a specific asset
def send_data(asset):
    asset_df = df[df['asset'] == asset]
    for _, record in asset_df.iterrows():
        producers[asset].send(KAFKA_TOPIC, record.to_dict())
        print(f"Sent from {asset}B: {record.to_dict()}")
        time.sleep(1)  # Simulate real-time streaming

# Start a thread for each asset
threads = []
for asset in unique_assets:
    thread = threading.Thread(target=send_data, args=(asset,))
    threads.append(thread)
    thread.start()

# Wait for all threads to finish
for thread in threads:
    thread.join()

print("Finished sending all data.")

# Close all producers
for producer in producers.values():
    producer.close()
```
